```python
import pandas as pd
def calculate_degree(row):
    
    
    row = row[:-1]
    listRow = row.split(":")
    student_name = listRow[0]
    grades = listRow[1].split(",")
    
    grade_1 = int(grades[0])
    grade_2 = int(grades[1])
    grade_3 = int(grades[2])
    
    average = (grade_1 + grade_2 + grade_3) / 3
    
    if average >= 90 and average <= 100:
        grade = "AA GEÇTİ"
    elif average >= 85 and average <= 89:
        grade = "BA GEÇTİ"
    elif average >= 80 and average <= 84:
        grade = "BB GEÇTİ"
    elif average >= 75 and average <= 79:
        grade = "CB GEÇTİ"
    elif average >= 70 and average <= 74:
        grade = "CC GEÇTİ"
    elif average >=65 and average <= 69:
        grade = "DC KOŞULLU GEÇTİ"
    elif average >= 60 and average <= 64:
        grade = "DD KOŞULLU GEÇTİ"
    elif average >= 50 and average <= 59:
         grade = "FD KALDI"
    else:
         grade = "FF KALDI"
    
    return student_name + ": " + grade + "\n"
    
  
def read_average():
    with open("ogrenci_notları.txt","r",encoding="utf-8") as file:
        for row in file:
            print(calculate_degree(row))

def add_degree():
    name = input("ÖĞRENCİNİN ADI: ")
    surname = input("ÖĞRENCİNİN SOYADI: ")
    grade_1 = input("ÖĞRENCİNİN 1.SINAVI: ")
    grade_2 = input("ÖĞRENCİNİN 2.SINAVI: ")
    grade_3 = input("ÖĞRENCİNİN 3.SINAVI: ")
    
    with open("ogrenci_notları.txt", "a", encoding="utf-8") as file:
        file.write(name + " " + surname + ":" + grade_1 + "," + grade_2 + "," + grade_3 + "\n")
    
def save_degree():
    with open("ogrenci_notları.txt", "r" ,encoding="utf-8") as file:
        listRows = []
        for i in file:
            listRows.append(calculate_degree(i))
            
    with open ("harf_notları.txt", "w", encoding="utf-8") as file2:
        for i in listRows:
            file2.write(i)
            
def frame():
    with open("harf_notları.txt", "r", encoding="utf-8") as file:
        data = pd.DataFrame(file)
        print(data)
    
def excel():
    with open("harf_notları.txt", "r" ,encoding="utf-8") as file:
        data = pd.DataFrame(file)
        data.to_excel("excel_harfNotları.xlsx")
    
    
            
    
    
print("************ ÖĞRENCİ NOT KAYIT FORMUNA HOŞ GELDİNİZ ******************")
while True:
    choosen = input("1- Ogrenci Girisi\n2- Notları Gör\n3- Notları Kaydet\n4- Data Frame Şeklinde Göster\n5- Excele Dönüştür\n6- Çıkış Yap\n\n")
    
    if int(choosen) == 1:
        note_entry()

    elif int(choosen) == 2:
        read_average()
        

    elif int(choosen) == 3:
        save_degree()
        print("NOTLAR KAYDEDİLDİ")
        
    elif int(choosen) == 4:
        frame()
        
    elif int(choosen) == 5:
        excel()
        
        

    else:
        print("SİSTEM SONLANDIRILDI.")
        break
```
